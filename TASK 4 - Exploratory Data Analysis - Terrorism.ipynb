Objective: Performing Exploratory Data Analysis on the dataset "Global Terrorism", hot zones can be located and as security or defense analysts these zones can be rectified.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=15,11
```

```python
# Loading the dataset
df=pd.read_csv("globalterrorismdb_0718dist.csv",encoding='latin-1')
```

```python
# Checking for missing data
df.isnull().sum()
```

```python
# Percentage of missing data
df.isnull().sum()/len(df)*100
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='rainbow_r')
```

```python
# Dropping the values
df.dropna(axis=1,inplace=True)
```

```python
# Checking the shape after droping null values
df.shape
```

```python
# Correlation
df.corr()
```

```python
sns.countplot(x=df['iyear'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x=df['iyear'], hue='success', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x=df['region_txt'], hue='success', data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country_txt'].value_counts()[:20].values)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=df['attacktype1_txt'].value_counts()[:20].index,y=df['attacktype1_txt'].value_counts()[:20].values)
plt.ylabel('Total Attack', fontsize=15)
plt.xticks(rotation=90)
plt.xlabel('Attack Type', fontsize=15)
plt.show()
```

```python
pd.crosstab(df.iyear, df.region).plot(kind='area')
```

```python
attack = df.country_txt.value_counts()[:10]
attack
```

```python
df.gname.value_counts()[1:10]
```

```python
sns.barplot(df['country_txt'].value_counts()[:10].index,df['country_txt'].value_counts()[:10].values,palette='YlOrBr_r')
plt.title('Top Countries Affected')
```

```python
df['attacktype1_txt'].value_counts().plot(kind='bar',color='magenta')
plt.xticks(rotation = 50)
plt.title("Name of attacktype")
```

```python
df['gname'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='green')
plt.title("Top 10 terrorist group attack",fontsize=20)
```
